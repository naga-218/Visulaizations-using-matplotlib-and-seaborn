## Visualization techniques using matplotlib

### Part -1

Matplotlib architecture is composed of 3 main layers:-

The Back — End Layer:- made up of the FigureCanvas, Renderer, Event.

The Artist Layer:- This is where much of the heavy lifting happens. Such as for writing a program or UI application to be shared with other developers

The Scripting Layer:- Made up of pyplot, as well as the layer for everyday purposes. This layer is considered a lighter scripting interface to simplify common tasks and for a quick and easy generation of graphics and plots.

##### The Back — End Layer:

As its name implies, the Back-End layer is the default layer that supports the other layers. It has 3 built-in abstract interface classes:-

FigureCanvas: This is the matplotlib.backend_bases.FigureCanvas object: It encompasses the area on which the figure or image is drawn.

Renderer: This is the matplotlib.backend_bases.Renderer object: It knows how to draw on the figure canvas.

Event: matplotlib.backend_bases.Event: This handles user inputs such as keyboard strokes and mouse clicks

#### The Artist Layer:

Comprised of One Main Object — FigureArtist:- The Artist knows how to use the renderer to draw and put ink on the FigureCanvas. Everything we see in a Matplotlib figure is an Artist instance. Titles, lines, tick labels and images all correspond to individual Artist instances.

Types of Artist Objects:

Primitive: Such as Lines, Rectangles, Circles and Texts.

Composite: Such as Axis, Ticks, Axes, and Figure.

#### The Scripting Layer:

The previous Artist layer is quite heavy on syntax and clearly for professional developers. Therefore the Scripting layer was developed for scientists who are not professional programmers. Clearly a lot of people prefer the Scripting Layer to the more syntax-heavy Artist layer, for quick analysis of some data.

Matplotlib Scripting Layer is essentially the all popular matplotlib.pyplot interface, which automates the process of defining a FigureCanvas, and defining a FigureArtist Instance and connecting them.

### 2. Area plot

```python
df['Total'] = df.loc[:, years].sum(axis = 1)
df.head()
```

```python
df.sort_values('Total', ascending=False, inplace=True)
df.head()
```

```python
top_5_df = df.head()
```

```python
top_5_df = top_5_df[years].T
top_5_df.head()
```

```python
dict_ = {'fontsize':18, 'fontweight':10, 'verticalalignment':'baseline',
       'color':'Red'}


top_5_df.plot(kind = 'area', figsize = (10, 6))
plt.title('Area Plot of Top 5 Migrating Countries to Canada', fontdict=dict_)
dict_['fontsize'] = 14
plt.xlabel('Years', fontdict = dict_)
plt.ylabel('Immigrants', fontdict=dict_)
plt.show()
```

```python
#col = sns.color_palette("hls", 5)
```

```python
# plt.plot([], [], color='blue', label='India')
# plt.plot([], [], color='orange', label='China')
# plt.plot([], [], color='brown', label='United Kingdom of Great Britain and Northern Ireland')
# plt.plot([], [], color='green', label='Philippines')
# plt.plot([], [], color='red', label='Pakistan')

plt.stackplot(years, [top_5_df['India'],
                               top_5_df['China'],
                               top_5_df['United Kingdom of Great Britain and Northern Ireland'],
                               top_5_df['Philippines'],
                               top_5_df['Pakistan']],
              labels= ['India', 'China', 'United Kingdom of Great Britain and Northern Ireland',
                      'Philippines', 'Pakistan'],alpha = 0.8)

plt.legend(loc=2, fontsize = 'large')
plt.show()
```

#### Let’s re-plot the above Area plot as an unstacked plot

```python
top_5_df.plot(kind = 'area', figsize = (10, 6), stacked = False)
plt.title('Area Plot of Top 5 Migrating Countries to Canada', fontdict=dict_)

dict_['fontsize'] = 14
plt.xlabel('Years', fontdict=dict_)
plt.ylabel('Immigrants', fontdict=dict_)
plt.show()
```

```python
fig = plt.figure(figsize=(20,6)) # create figure

ax0 = fig.add_subplot(121) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(122) # add subplot 2 (1 row, 2 columns, second plot)

# Subplot 1: Stacked Area plot
top_5_df.plot(kind='area', alpha=0.4, ax=ax0) # add to subplot 1
ax0.set_title('Stacked Area Plot with alpha=0.4')
ax0.set_ylabel('Immigrants')
ax0.set_xlabel('Years')

# Subplot 2: Unstacked Area Plot
top_5_df.plot(kind='area', stacked=False, alpha=0.4, ax=ax1) # add to subplot 2
ax1.set_title ('Unstacked Area plot with alpha=0.4')
ax1.set_ylabel('Immigrants')
ax1.set_xlabel('Years')

plt.show()
```

#### Use the Artist layer to display the immigration distribution for the top 5 Countries to Canada for the years 1980–2013? Use an overlapping stacked plot with 15 bins and a transparency value of 0.5.

```python
top_5 = df.sort_values(by = 'Total', ascending=False).head()
top_5 = top_5.loc[top_5.index, years].T
top_5.head()
```

```python
top_5.rename(columns={'United Kingdom of Great Britain and Northern Ireland':'United_Kingdom(GB/NI)'}, inplace=True)
```

```python
count, bin_edges = np.histogram(top_5, 15)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1)

top_5.plot.hist(xticks=bin_edges,
               stacked = True,
               alpha = 0.5,
               color = ['orange', 'brown', 'mediumpurple', 'tomato', 'black'],
               figsize = (10, 6),
               bins = len(count),
               ax = ax)


ax.set_title('Histplot of migrations from Top-5 countries to Canada (1980 - 2013)',fontstyle='italic')
ax.set_ylabel('Number of Years')
ax.set_xlabel('Number of Immigrants')
plt.show()
```
